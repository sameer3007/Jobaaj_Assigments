```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wr
```

```python
from sklearn.datasets import load_iris
data=load_iris()
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df.head()
```

```python
#standarisdation
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(df)
x_scaled
```

```python
#pca
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_x_scaled=pca.fit_transform(x_scaled)
pca_x_scaled
```

```python
plt.scatter(pca_x_scaled[:,0],pca_x_scaled[:,1],c=data.target)
```

```python
# agglomerative clustring
# to construct a dendogram
import scipy.cluster.hierarchy as sc
sc.dendrogram(sc.linkage(pca_x_scaled,method='ward'))
plt.title('dendrogram')
plt.xlabel('sample index')
plt.ylabel('eucledian distance')
```

# agglomerative hierarichal clustring algo

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
cluster.fit(pca_x_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_x_scaled[:,0],pca_x_scaled[:,1],c=cluster.labels_)
```

# K-means clustring VS hierarichal clustring(AgglomerativeClustering)
